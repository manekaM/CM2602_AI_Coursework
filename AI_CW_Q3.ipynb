#Task 01

```python
#TASK 1

ROWS = 6
COLS = 6

def node_to_xy(node_id):
    return node_id // ROWS, node_id % ROWS

start_node = random.randint(0, 11)
goal_node = random.randint(24, 35)
remaining = [n for n in range(36) if n not in (start_node, goal_node)]
barrier_nodes = random.sample(remaining, 4)

print(f"Start: {start_node}, Goal: {goal_node}, Barriers: {sorted(barrier_nodes)}\n")

CELL = 6  # fixed cell width

# Column headers
print(" " * 6, end="")
for col in range(COLS):
    print(f"{col:^{CELL}}", end="")
print()

divider = "      +" + ("-----+" * COLS)

print(divider)
for row in range(ROWS):
    print(f"  {row}   |", end="")
    for col in range(COLS):
        node_id = col * ROWS + row
        if node_id == start_node:
            label = f"S{node_id}"
        elif node_id == goal_node:
            label = f"G{node_id}"
        elif node_id in barrier_nodes:
            label = f"X{node_id}"
        else:
            label = str(node_id)
        print(f"{label:^5}|", end="")
    print()
    print(divider)

print("\nLegend:  S = Start   G = Goal   X = Barrier")
print(f"\nNode coordinates (x=col, y=row):")
print(f"  Start   node {start_node:>2} -> {node_to_xy(start_node)}")
print(f"  Goal    node {goal_node:>2} -> {node_to_xy(goal_node)}")
for b in sorted(barrier_nodes):
    print(f"  Barrier node {b:>2} -> {node_to_xy(b)}")
```

#Task 2

```python
#TASK 2

#Helpers (reuse maze variables from previous cell)
def node_to_xy(node_id):
    return node_id // ROWS, node_id % ROWS

def get_neighbors(node_id):
    x, y = node_to_xy(node_id)
    neighbors = []
    for dx in [-1, 0, 1]:
        for dy in [-1, 0, 1]:
            if dx == 0 and dy == 0:
                continue
            nx, ny = x + dx, y + dy
            if 0 <= nx < COLS and 0 <= ny < ROWS:
                nid = xy_to_node(nx, ny)
                if nid not in barrier_nodes:
                    neighbors.append(nid)
    return sorted(neighbors)

def edge_cost(a, b):
    ax, ay = node_to_xy(a)
    bx, by = node_to_xy(b)
    return math.sqrt((ax - bx)**2 + (ay - by)**2)

# IDDFS
visited_log = []

def dls(node, goal, depth, path):
    visited_log.append(node)
    if node == goal:
        return True
    if depth == 0:
        return False
    for neighbor in get_neighbors(node):
        if neighbor not in path:
            path.append(neighbor)
            if dls(neighbor, goal, depth - 1, path):
                return True
            path.pop()
    return False

def iddfs(start, goal):
    global visited_log
    visited_log = []
    for depth_limit in range(36):
        path = [start]
        if dls(start, goal, depth_limit, path):
            return path, depth_limit
    return None, -1

# Print results
path, depth = iddfs(start_node, goal_node)

print("ITERATIVE DEEPENING DFS")

print(f"\nVisited nodes ({len(visited_log)} total):")
print(" -> ".join(map(str, visited_log)))

print(f"\nTime to find goal : {len(visited_log)} minute(s)")

if path:
    print(f"\nFinal path (depth {depth}):")
    print(" -> ".join(map(str, path)))

    total_cost = sum(edge_cost(path[i], path[i+1]) for i in range(len(path)-1))
    print(f"\nPath edge costs (Euclidean distance):")
    for i in range(len(path)-1):
        cost = edge_cost(path[i], path[i+1])
        move = "diagonal" if cost > 1.0 else "straight"
        print(f"  Node {path[i]:>2} -> {path[i+1]:>2} : {cost:.4f}  ({move})")
    print(f"  Total path cost : {total_cost:.4f}")
else:
    print("\nNo path found.")
```

#Task 3

```python
#TASK 3


# Calculates the heuristic cost for a node using Chebyshev Distance.
# Formula: d(N, G) = max(|Nx - Gx|, |Ny - Gy|)

def calculate_heuristic_cost(current_node, goal_node):
    # Get the (x, y) coordinates for the current node and the goal node
    nx, ny = node_to_xy(current_node)
    gx, gy = node_to_xy(goal_node)

    # Calculate the absolute differences
    dx = abs(nx - gx)
    dy = abs(ny - gy)

    # Return the maximum of the two differences
    return max(dx, dy)

#Heuristic cost of all the 36 nodes
print(f"Goal Node is: {goal_node}\n")

for node in range(36):
    cost = calculate_heuristic_cost(node, goal_node)
    print(f"Heuristic cost for Node {node:>2}: {cost}")
```

#Task 4

```python
#TASK 4
def best_first_search(start, goal):
    # Priority queue stores tuples: (heuristic_cost, node_id)
    open_list = [(calculate_heuristic_cost(start, goal), start)]

    # Dictionary to track paths for reconstruction: {node: parent_node}
    came_from = {start: None}

    # List to track the exact order nodes are explored
    visited_log = []

    while open_list:
        # Pop the node with the lowest heuristic cost
        current_h, current_node = heapq.heappop(open_list)

        # Record the node as visited
        visited_log.append(current_node)

        # Check if we've reached the goal
        if current_node == goal:
            # Reconstruct the path backwards from goal to start
            path = []
            temp = current_node
            while temp is not None:
                path.append(temp)
                temp = came_from[temp]
            path.reverse() # Reverse to get start -> goal
            return path, visited_log

        # Process neighbors (get_neighbors already returns them sorted ascendingly)
        for neighbor in get_neighbors(current_node):
            if neighbor not in came_from: # Only add unvisited nodes
                came_from[neighbor] = current_node
                h_cost = calculate_heuristic_cost(neighbor, goal)
                heapq.heappush(open_list, (h_cost, neighbor))

    return None, visited_log # Return if no path is found

#Print the search algorithm performed
bfs_path, bfs_visited = best_first_search(start_node, goal_node)

print("BEST FIRST SEARCH (BFS)")

print(f"\nVisited nodes ({len(bfs_visited)} total):")
print(" -> ".join(map(str, bfs_visited)))

# It takes 1 minute to explore a single node.
print(f"\nTime to find goal : {len(bfs_visited)} minutes")

if bfs_path:
    print(f"\nFinal path:")
    print(" -> ".join(map(str, bfs_path)))
else:
    print("\nNo path found.")
```

#Task 6

```python
#TASK 6

# Three New Distance Measurements
def heuristic_manhattan(node, goal):
    #Manhattan Distance: d = |dx| + |dy|
    nx, ny = node_to_xy(node)
    gx, gy = node_to_xy(goal)
    return abs(nx - gx) + abs(ny - gy)

def heuristic_euclidean(node, goal):
    #Euclidean Distance: d = sqrt(dx^2 + dy^2)
    nx, ny = node_to_xy(node)
    gx, gy = node_to_xy(goal)
    return math.sqrt((nx - gx)**2 + (ny - gy)**2)

def heuristic_octile(node, goal):
    #Octile Distance: Tailored for 8-way grid movement
    nx, ny = node_to_xy(node)
    gx, gy = node_to_xy(goal)
    dx, dy = abs(nx - gx), abs(ny - gy)
    return max(dx, dy) + (math.sqrt(2) - 1) * min(dx, dy)

# Updated BFS to accept a custom heuristic function
def best_first_search_custom(start, goal, heuristic_func):
    open_list = [(heuristic_func(start, goal), start)]
    came_from = {start: None}
    visited_log_custom = []

    while open_list:
        current_h, current_node = heapq.heappop(open_list)
        visited_log_custom.append(current_node)

        if current_node == goal:
            path = []
            temp = current_node
            while temp is not None:
                path.append(temp)
                temp = came_from[temp]
            path.reverse()
            return path, visited_log_custom

        for neighbor in get_neighbors(current_node):
            if neighbor not in came_from:
                came_from[neighbor] = current_node
                h_cost = heuristic_func(neighbor, goal)
                heapq.heappush(open_list, (h_cost, neighbor))

    return None, visited_log_custom


print("REPEATING TASK 5 FOR 3 DISTANCE MEASUREMENTS")
print(f"\n")

heuristics = {
    "Manhattan": heuristic_manhattan,
    "Euclidean": heuristic_euclidean,
    "Octile": heuristic_octile
}

# Generating 3 new random mazes per heuristic(repatition of task 5)
for name, h_func in heuristics.items():
    print(f"\n" + "-" * 45)
    print(f" TESTING {name.upper()} DISTANCE ")
    print("-" * 45)

    bfs_times = []
    bfs_lengths = []

    for run in range(1, 4):
        # Setup a brand new Random Maze
        start_node = random.randint(0, 11)
        goal_node = random.randint(24, 35)
        remaining = [n for n in range(36) if n not in (start_node, goal_node)]

        global barrier_nodes
        barrier_nodes = random.sample(remaining, 4)

        # Run Best First Search with the current heuristic
        path, visited = best_first_search_custom(start_node, goal_node, h_func)
        time_taken = len(visited)
        length = sum(edge_cost(path[i], path[i+1]) for i in range(len(path)-1)) if path else 0

        bfs_times.append(time_taken)
        bfs_lengths.append(length)

        print(f"  Run {run} -> Start: {start_node:<2}, Goal: {goal_node:<2}, Barriers: {sorted(barrier_nodes)}")
        print(f"         -> Time: {time_taken:<4} mins | Path Cost: {length:.4f}")

    # Calculate Mean and Variance for the current heuristic
    print(f"\n  [ {name} Statistical Results ]")
    print(f"  Mean Time     : {statistics.mean(bfs_times):.2f} minutes")
    print(f"  Variance Time : {statistics.variance(bfs_times) if len(bfs_times) > 1 else 0:.2f}")
    print(f"  Mean Cost     : {statistics.mean(bfs_lengths):.4f}")
    print(f"  Variance Cost : {statistics.variance(bfs_lengths) if len(bfs_lengths) > 1 else 0:.4f}")
```
